```python
# Define the mapping function
def convert_duration(duration:pd.DataFrame):
    duration=ba['Duration'].to_list()
    duration_numer=[]
    for dc in duration:
        if dc.startswith('More') or dc.startswith('Less'):
            m=dc.split()
            duration_numer.append(int(m[2])+1)
        else:
            m=dc.split()
            j=m[0].split('-')
            duration_numer.append(int(j[0])+int(j[1])//2)

    #duration_numeric=np.array(duration_numeric).reshape(1,10)


    return duration_numer
```

```python
m=convert_duration(ba['Duration'])
j=pd.DataFrame({'Duration_numer':m})
ba=pd.concat([ba,j],axis=1)
display(ba.head())
```

```python
print('Average Investment Duration is: ',int(ba['Duration_numer'].mean()),' years',sep='')
```
